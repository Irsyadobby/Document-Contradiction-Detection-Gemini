# Read File PDF

```python
import pandas as pd
import numpy as np
import glob
import os
import re
import pdfplumber
```

```python
# Function membaca Jenis, Nomor, dan Tahun dokumen

def info_dokumen(nama_dokumen) :
    if nama_dokumen.startswith("Inpres") or nama_dokumen.startswith("Keppres") or nama_dokumen.startswith("Perpres") :
        jenis = "Peraturan Presiden"
    elif nama_dokumen.startswith("Kepmen") or nama_dokumen.startswith("PM") :
        jenis = "Peraturan Mentri"
    elif nama_dokumen.startswith("PP") :
        jenis = "Peraturan Pemerintah"
    elif nama_dokumen.startswith("UU") :
        jenis = "Undang-Undang"

    nomor = None 
    tahun = None

    nomor_tahun = re.findall(r"\d+", nama_dokumen)
    if len(nomor_tahun) >= 2 :
        nomor = int(nomor_tahun[0])
        tahun = int(nomor_tahun[1])

    return jenis, nomor, tahun 
```

```python
path = glob.glob("C:/Users/Irsyadobby/Documents/BDC 2023/Semifinal/Dataset/Peraturan Investasi/*.pdf")
path = [s.replace('\\','/') for s in path]

data = pd.DataFrame(columns = ['Jenis','Nomor','Tahun','Halaman','Teks'])

for files in path :
    # Read PDF  
    reader = pdfplumber.open(files)
    
    # Baca Informasi Dokumen 
    nama_dokumen = os.path.basename(files)

    for i in range(0,len(reader.pages)) :
        hal = i + 1
        page = reader.pages[i]
        text = page.extract_text()
        jenis, nomor, tahun = info_dokumen(nama_dokumen)
        data = data.append({'Jenis' : jenis, 'Nomor' : nomor, 'Tahun' : tahun, 'Halaman' : hal, 'Teks' : text}, ignore_index = True)
```

```python
data.sort_values(['Tahun', 'Halaman'], ascending = [True, True], ignore_index = True, inplace = True)
```

# Pre-Processing Data

```python
import string
import re 
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
```

```python
# lowercase string
def lowercase_string(text):
    return text.lower()
data['Teks'] = data['Teks'].apply(lowercase_string)

# remove punctuation 
def remove_punctuation(text):
    return text.translate(str.maketrans("","",string.punctuation))
data['Teks'] = data['Teks'].apply(remove_punctuation)

# remove multiple whitespace
def remove_whitespace_multiple(text):
    return ' '.join(text.split())
data['Teks'] = data['Teks'].apply(remove_whitespace_multiple)

# remove mention and special character
def mention(text):
    return re.sub(r"(?:\@|https?\://)\S+", "", text)
data['Teks'] = data['Teks'].apply(mention)

# remove single char
def remove_singl_char(text):
    return re.sub(r"\b[a-zA-Z]\b", "", text)
data['Teks'] = data['Teks'].apply(remove_singl_char)
```

```python
# Tokenize 
def word_tokenize_wrapper(text):
    return word_tokenize(text)
data['Tokens'] = data['Teks'].apply(word_tokenize_wrapper)
```

```python
# remove stopwords
stopwords = pd.DataFrame(['abcdefghijkl','www.hukumonline.com','wwwhukumonlinecom'])
stop = stopwords[0].values.tolist()

def stopwords_removal(words):
    return [word for word in words if word not in stop]
data['Tokens'] = data['Tokens'].apply(stopwords_removal)
```
